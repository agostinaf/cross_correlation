```python
import pandas as pd
from bokeh.plotting import output_notebook, figure, ColumnDataSource, show
from bokeh.models.tools import HoverTool
from bokeh.models.glyphs import Circle, Ellipse

output_notebook()
```

```python
df2 = pd.read_csv('./catalogs/FGLxSFR.csv')
df1 = df2.drop_duplicates(subset=['ID1'])
```

```python
source1 = ColumnDataSource(data=df1)
source2 = ColumnDataSource(data=df2)

TOOLTIPS1 = [
    ("RA, DE", "@RA1, @DE1"),
    ("ID1", "@ID1"),
    ("NAME", "@NAME1"),
]

TOOLTIPS2 = [
    ("RA, DE", "@RA2, @DE2"),
    ("NAME", "@NAME2"),
    ("ID2", "@ID2"),
]

p = figure(plot_width=600, plot_height=400)
c1 = Circle(x='RA1', y='DE1', size=3, line_color=None, fill_color="#1f77b4", fill_alpha=0.9)
c3 = Ellipse(x='RA1', y='DE1', width='SIGMA_RA1', height='SIGMA_DE1', line_color='gray', line_alpha=0.7, fill_color=None)
c2 = Circle(x='RA2', y='DE2', size=3, line_color=None, fill_color="#ff7f0e", fill_alpha=0.7)
c4 = Ellipse(x='RA2', y='DE2', width='SIGMA_RA2', height='SIGMA_DE2', line_color='gray', line_alpha=0.7, fill_color=None)

hover1 = HoverTool(tooltips=TOOLTIPS1, names = ['cat1'])
hover2 = HoverTool(tooltips=TOOLTIPS2, names = ['cat2'])

p.add_glyph(source1, c1, name='cat1')
p.add_glyph(source2, c2, name='cat2')
p.add_glyph(source1, c3)
p.add_glyph(source2, c4)
p.add_tools(hover1)
p.add_tools(hover2)

show(p)
```
